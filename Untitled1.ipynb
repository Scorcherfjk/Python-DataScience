```python
%matplotlib inline
import matplotlib.pyplot as plt
from random import sample
```

```python
gastos = [50,120,80,95,100]
gastos2 = [30,20,100,70,200]
meses = ["enero","febrero","marzo","abril","mayo"]
```

```python
posiciones = [0,1,2,3,4]

plt.title("grafica de gastos")
plt.ylabel("gastos")
plt.xlabel("meses")
plt.xticks(posiciones,meses)
plt.ylim(0,150)

plt.plot(gastos, 
         label="gastos", 
         ls="-.",
         lw=3,
         color="red",
         alpha=1,
         marker="*",
         markersize=15,
         markerfacecolor="blue",
         markeredgecolor="green",
         markeredgewidth=1)

plt.plot(gastos2, 
         label="gastos2", 
         ls="--")

plt.legend(loc=4)
```
