### Obteniendo los datos del archivo

```python
iris = open("iris.csv","r")
lineas = iris.readlines()
iris.close()
del(iris)
```

###### Verificamos...

```python
for linea in lineas:
    print(linea)
```

### Eliminamos el salto de linea que viene al final 

```python
lineaslimpias = []
for linea in lineas:
    
    # metodo 1 - funciones internas
    lineaslimpias.append(linea.strip())
    
    # metodo 2 - slicing
    # lineaslimpias.append(linea[0:-1])
```

###### Verificamos...

### AÃ±adimos cada linea a una lista formando una matriz

```python
listas = []
for linea in lineaslimpias:
    listas.append(linea.split(","))
```

###### Verificamos...

```python
for linea in listas:
    print(linea)
```

### Convertimos las columnas correspondientes a un tipo de dato util para calculos ( float )

```python
for linea in listas[1:]:
    linea[0] = float(linea[0])
    linea[1] = float(linea[1])
    linea[2] = float(linea[2])
    linea[3] = float(linea[3])
```

###### Verificamos...

###### Obligatorio abrir el archivo

```python
iris = open("iris.csv","r")
datos = iris.readlines()
iris.close()
del(iris)
datoslimpios = []
```

###### Limpieza de datos

```python
for linea in datos:
    datoslimpios.append(linea[0:-1].split(","))  
for linea in datoslimpios[1:]:
    linea[0] = float(linea[0])
    linea[1] = float(linea[1])
    linea[2] = float(linea[2])
    linea[3] = float(linea[3])
```

###### Verificamos...

```python
for linea in datoslimpios:
    print(linea)
```

### Declaramos las funciones estadisticas

Las funciones de maximo, minimo y conteo ya existen

```python
def mean(lista):
    a = 0
    for valor in lista:
        a += valor
    return a / len(lista)
```

### Dividimos los datos

```python
sepal_length = [] 
sepal_width = []
petal_length = [] 
petal_width = []
for linea in listas[1:]:
    sepal_length.append(linea[0])
    sepal_width.append(linea[1])
    petal_length.append(linea[2]) 
    petal_width.append(linea[3])
```

### Realizamos los calculos

```python
print("la media aritmetica de la longitud del sepalo es: {:.4f}".format(mean(sepal_length)))
print("el valor maximo de la longitud del sepalo es: {:.1f}".format(max(sepal_length)))
print("el valor minimo de la longitud del sepalo es: {:.1f}".format(min(sepal_length)))
print("el total de los elementos recolectados de la longitud del sepalo es: {:.0f}".format(len(sepal_length)))
```

###### repetir esto la cantidad de columnas existentes editando el resultado

# O sino

```python
def resultadosSimples(lista):
    print("+--------+--------+----------+--------+")
    print("| maximo | minimo | promedio | total  |")
    print("+--------+--------+----------+--------+")
    print("|   {:.1f}  |   {:.1f}  |  {:.4f}  |  {:04.0f}  |".format(max(lista),min(lista),mean(lista),len(lista)))
    print("+--------+--------+----------+--------+")
```

### Utilizamos la funcion y mostramos los resultados

```python
print("Sepal length")
resultadosSimples(sepal_length)

print("\nSepal width")
resultadosSimples(sepal_width)

print("\nPetal length")
resultadosSimples(petal_length)

print("\nPetal width")
resultadosSimples(petal_width)
```

# Asi finalizamos con los datos en general y empezamos con los especificos

### Dividimos los datos por tipo de flor

```python
setosa = []
versicolor = [] 
virginica = []
for linea in listas[1:]:
    if linea[4] == 'setosa':
        setosa.append(linea)
    elif linea[4] == 'versicolor':
        versicolor.append(linea)
    else:
        virginica.append(linea)
```

### Hacemos los calculos y preparamos el resultado

```python
def resultadosEspecificos(lista):
    sepal_length = [] 
    sepal_width = []
    petal_length = [] 
    petal_width = []
    for linea in lista:
        sepal_length.append(linea[0])
        sepal_width.append(linea[1])
        petal_length.append(linea[2]) 
        petal_width.append(linea[3])
    
    
    print("+------------+--------+--------+----------+--------+")
    print("| Dato       | Maximo | Minimo | Promedio | Total  |")
    print("+------------+--------+--------+----------+--------+")
    print("|sepal_length|   {:.1f}  |   {:.1f}  |  {:.4f}  |  {:04.0f}  |".format(max(sepal_length),
                                                                                  min(sepal_length),
                                                                                  mean(sepal_length),
                                                                                  len(sepal_length)))
    
    print("+------------+--------+--------+----------+--------+")
    print("|sepal_width |   {:.1f}  |   {:.1f}  |  {:.4f}  |  {:04.0f}  |".format(max(sepal_width),
                                                                                  min(sepal_width),
                                                                                  mean(sepal_width),
                                                                                  len(sepal_width)))
    
    print("+------------+--------+--------+----------+--------+")
    print("|petal_length|   {:.1f}  |   {:.1f}  |  {:.4f}  |  {:04.0f}  |".format(max(petal_length),
                                                                                  min(petal_length),
                                                                                  mean(petal_length),
                                                                                  len(petal_length)))
    
    print("+------------+--------+--------+----------+--------+")
    print("|petal_width |   {:.1f}  |   {:.1f}  |  {:.4f}  |  {:04.0f}  |".format(max(petal_width),
                                                                                  min(petal_width),
                                                                                  mean(petal_width),
                                                                                  len(petal_width)))
    
    print("+------------+--------+--------+----------+--------+")
```

### Mostramos los resultados

```python
print("Setosa")
resultadosEspecificos(setosa)

print("\nVersicolor")
resultadosEspecificos(versicolor)

print("\nVirginica")
resultadosEspecificos(virginica)
```
